# Visualize Decision Trees

## Requires graphviz library and dot files exported from model

```python
import os, sys, re
import numpy as np
import graphviz
from pathlib import Path
from sklearn import tree
import warnings
```

# Save trees to PDF images

```python
n_estimators = 100
min_samples_split = 20
tree_dot_dir = f'tree_dots/n{n_estimators}s{min_samples_split}'
tree_img_dir = f'tree_images/n{n_estimators}s{min_samples_split}'
Path(tree_img_dir).mkdir(exist_ok=True)

trees = np.arange(0, n_estimators, 20) + 10

for tree_num in trees:
    fn = f'{tree_dot_dir}/tree_dot_{tree_num}.txt'
    with open(fn, 'r') as f:
        dot_data = f.read()
    graph = graphviz.Source(dot_data)
    fn = f'{tree_img_dir}/tree_{tree_num}'
    fmt = 'pdf'
    graph.render(fn, format=fmt)
    if os.path.exists(fn):
        # cleanup non-pdf version of file
        os.remove(fn)
    print(f'{fn}.{fmt}')
```
