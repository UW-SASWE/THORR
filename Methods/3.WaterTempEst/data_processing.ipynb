```python
import pandas as pd
import glob
import os
from pathlib import Path
```

```python
proj_dir = Path("../..")
```

```python
file_names = [os.path.basename(x) for x in glob.glob(str(proj_dir) + "Data/RAT_outputs/dels/*.csv")]
```

```python
# file_names = [os.path.basename(x) for x in glob.glob(r"/Users/gdarkwah/Library/CloudStorage/OneDrive-UW/01-Research/01-Hydrothermal History/Methods/machine_learning/ML2/RAT_outputs/dels/*.csv")]
dels_file_names = [os.path.basename(x) for x in glob.glob(str(proj_dir) + "/Data/RAT_outputs/dels/*.csv")]
sarea_file_names = [os.path.basename(x) for x in glob.glob(str(proj_dir) + "/Data/RAT_outputs/sarea_tmsos/*.csv")]
GrandID = [int(x.split('_')[0]) for x in dels_file_names]

df = pd.DataFrame({'Grand_ID': GrandID, 'file_name': dels_file_names})
```

```python
# rat_data = list()
rat_dels = list()
rat_sarea = list()
```

```python
for index, row in df.iterrows():
    delS = pd.read_csv(
        str(proj_dir) + "/Data/RAT_outputs/dels/" + row["file_name"],
        parse_dates=["date"],
    )
    sarea = pd.read_csv(
        str(proj_dir) + "/Data/RAT_outputs/sarea_tmsos/" + row["file_name"],
        parse_dates=["date"],
    )
    delS.rename(columns={"date": "Date"}, inplace=True)
    sarea.rename(columns={"date": "Date"}, inplace=True)
    delS["Date"] = pd.to_datetime(delS["Date"])
    sarea["Date"] = pd.to_datetime(sarea["Date"])
    delS.set_index("Date", inplace=True)
    sarea.set_index("Date", inplace=True)

    # resample to semi-monthly
    delS = delS.resample("SMS").mean()
    sarea = sarea.resample("SMS").mean()

    delS["GranD_ID"] = int(row["Grand_ID"])
    sarea["GranD_ID"] = int(row["Grand_ID"])
    # print(row["GrandID"])
    # break

    # rat_data.append(delS)
    rat_dels.append(delS)
    rat_sarea.append(sarea)
```

```python
# rat_data = pd.concat(rat_data)
rat_dels = pd.concat(rat_dels)
rat_sarea = pd.concat(rat_sarea)
```

```python
rat_dels.to_csv("rat_dels.csv")
rat_sarea.to_csv("rat_sarea.csv")
```

## ResOps

```python
import pandas as pd
from pathlib import Path
import numpy as np
import geopandas as gpd
```

```python
proj_dir = Path("../../")
```

```python
reaches_and_dams = pd.read_csv(Path(proj_dir/"Methods/3.WaterTempEst/reaches_and_dams.csv"))
crb_dams = gpd.read_file(Path(proj_dir/"Data/GIS/shapefiles/CRBDams.shp"))
```

```python
resops_storage = pd.read_csv(Path(proj_dir, "Data/ResOpsUS/time_series_single_variable_table/DAILY_AV_STORAGE_MCM.csv"))
resops_inflow = pd.read_csv(Path(proj_dir, "Data/ResOpsUS/time_series_single_variable_table/DAILY_AV_INFLOW_CUMECS.csv"))
resops_outflow = pd.read_csv(Path(proj_dir, "Data/ResOpsUS/time_series_single_variable_table/DAILY_AV_OUTFLOW_CUMECS.csv"))

resops_timeseries_inventory = pd.read_csv(Path(proj_dir, "Data/ResOpsUS/attributes/time_series_inventory.csv"))
```

```python
resops_storage["date"] = pd.to_datetime(resops_storage["date"])
resops_inflow["date"] = pd.to_datetime(resops_inflow["date"])
resops_outflow["date"] = pd.to_datetime(resops_outflow["date"])
```

```python
resops_storage.replace('NaN', np.nan, inplace=True)
```

```python
reservoir_dynamics_list = list()
```

```python
for dam_id in crb_dams["GRAND_ID"]:
    if dam_id not in resops_timeseries_inventory["DAM_ID"].unique():
        # print("Dam ID: " + str(dam_id) + " not in ResOpsUS inventory")
        continue
    else:
        storage = resops_storage[["date", str(dam_id)]].dropna().copy()
        storage.rename(columns={str(dam_id): "storage"}, inplace=True)
        inflow = resops_inflow[["date", str(dam_id)]].dropna().copy()
        inflow.rename(columns={str(dam_id): "inflow"}, inplace=True)
        outflow = resops_outflow[["date", str(dam_id)]].dropna().copy()
        outflow.rename(columns={str(dam_id): "outflow"}, inplace=True)

        # storage.set_index("date", inplace=True)
        # inflow.set_index("date", inplace=True)
        # outflow.set_index("date", inplace=True)

        # merge the dataframes on date
        df = pd.merge(storage, inflow, on="date")
        df = pd.merge(df, outflow, on="date")

        df["storage"] = df["storage"].astype(float)
        df["inflow"] = df["inflow"].astype(float)
        df["outflow"] = df["outflow"].astype(float)
        df["GranD_ID"] = int(dam_id)

        reservoir_dynamics_list.append(df)
```

```python
reservoir_dynamics_df = pd.concat(reservoir_dynamics_list)
reservoir_dynamics_df.rename(columns={"date": "Date"}, inplace=True)
```

```python
reservoir_dynamics_df.to_csv(Path(proj_dir, "Methods/3.WaterTempEst/resops.csv"), index=False)
```
