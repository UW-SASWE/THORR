```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
proj_dir = Path('../..')
```

```python
gc_forebay_fn = proj_dir / 'Data/insitu/conditions/processed/DART_GCL.csv'
gc_tailrace_fn = proj_dir / 'Data/insitu/conditions/processed/DART_GCGW.csv'
```

```python
gc_forebay_data = pd.read_csv(gc_forebay_fn, parse_dates=True)
gc_tailrace_data = pd.read_csv(gc_tailrace_fn, parse_dates=True)
```

```python
gc_forebay_data['date'] = pd.to_datetime(gc_forebay_data['date'])
gc_tailrace_data['date'] = pd.to_datetime(gc_tailrace_data['date'])

# get month and year
gc_forebay_data['month'] = gc_forebay_data['date'].dt.month
gc_forebay_data['year'] = gc_forebay_data['date'].dt.year
gc_forebay_data['doy'] = gc_forebay_data['date'].dt.dayofyear

gc_tailrace_data['month'] = gc_tailrace_data['date'].dt.month
gc_tailrace_data['year'] = gc_tailrace_data['date'].dt.year
gc_tailrace_data['doy'] = gc_tailrace_data['date'].dt.dayofyear
```

```python
# find the monthly averages for the forebay and tailrace
gc_forebay_monthly = gc_forebay_data.groupby(['month']).mean()
gc_tailrace_monthly = gc_tailrace_data.groupby(['month']).mean()

# find the daily averages for the forebay and tailrace
gc_forebay_daily = gc_forebay_data.groupby(['doy']).mean()
gc_tailrace_daily = gc_tailrace_data.groupby(['doy']).mean()
```
